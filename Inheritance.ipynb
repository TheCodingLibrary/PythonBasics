# Inheritance in Python - Classes and OOP

# 1. What is inheritance and why use it? 

# 2. Create a Parent Class 

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def printNameAndAge(self):
        name = self.name
        age = self.age
        print("My name is", name, "and i'm", age)

        
person = Person("Jack", 23)
person.printNameAndAge()
```

# 3. Create a Child Class

```python
class Student(Person):
    def __init__(self, name, age):
        Person.__init__(self, name, age)
        
        
student = Student("Jackie", 22)
student.printNameAndAge()
```

# 4. Add an Attribute to the Child Class

```python
class Student(Person):
    def __init__(self, name, age, grades):
        # Super function will automatically inherit all met
        super().__init__(name, age)
        self.grades = grades
    
    def printGrades(self):
        grades = self.grades
        print("Grades:", grades)
        
student = Student("Jackie", 22, ["A", "A", "B"])
student.printNameAndAge()
student.printGrades()
```
