# File Handling in Python 

# 1. Open and Create a File 

# Function - open(filename, mode) 

```python
# Create a text file
file = open("testFile.txt", "x")
```

# 2. Read a File

```python
# Read file - specify path
file = open("testFile.txt", "r")
textFromFile = file.read()

print(textFromFile)
```

```python
# Read only parts from a file
file = open("testFile.txt", "r")
part = file.read(2)

print(part)
```

```python
# Read a Line
file = open("testFile.txt", "r")
firstLine = file.readline()
secondLine = file.readline()

print("First: ", firstLine)
print("Second: ", secondLine)
```

```python
# Close the file when done using it
file = open("testFile.txt", "r")
file.close()
file.read()
```

# 3. Write to a File 

```python
# Write to a file
file = open("testFile.txt", "a")
file.write("\nI just wrote this line to the file\n")
file.close()

# Read the file after writing to it
file = open("testFile.txt", "r")
fileRead = file.read()
print(fileRead)
file.close()
```

# 4. Delete a File 

```python
# Remove a file
import os
os.remove("testFile.txt")
```

```python
# Check if a file exist in path
if os.path.exists("testFile.txt"):
    os.remove("testFile.txt")
else:
    print("The file doesn't exist!")
```
