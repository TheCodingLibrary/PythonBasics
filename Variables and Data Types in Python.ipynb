# Variables and Data Types in Python

# 1. Variables 

```python
# This is how to make a comment
a = 2
b = 3
c = a + b

# Print the result
print(c)
```

# 2. Global and Local Variables


```python
d = 4
e = 3

# Add two numbers
def add(num1, num2):
    result = num1 + num2
    return result

f = add(d,e)

print(f)
```

# 3. Assign Values to Multiple Variables 

```python
x, y, z = 1, 2, 3

print(x, y, z)
```

# 4. Data Types 

```python
stringName = "Welcome"
intName = 1
floatName = 0.5
listName = ["orange", "green", "blue"]
tupleName = ("orange", "green", "blue")
rangeName = range(3)

print(tupleName)
print(type(intName))
```
